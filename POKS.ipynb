```python
## List is a collection which is ordered and changeable. Allows duplicate members.Goes with square brackets or list() constructor. 
## You can access items in a list by referring to the index number.You can loop through the list items by using a for loop.
## Use if statement to check if an item exists in the list.

mylist = ["apple", "banana", "cherry", "date",24, True, 24.5] 
item = mylist[2]
print(item)  # Output: banana
for m in mylist:
    print(m)

if False in mylist:
     print("Yes")
else:
        print("No")

## Check how many items are in the list, use the len() function:
print(len(mylist))

## Add items to a list, use the append() method to add an item to the end of the list
mylist.append("elderberry")
print(mylist)

## Use the insert() method to add an item at the specified index:
mylist.insert(1, "blueberry")
print(mylist)

## Use pop() method to remove an item from the end of the list:
mylist.pop()
print(mylist.pop())
print(mylist)

## Use the remove() method to remove a specified item:
mylist.remove("banana")
print(mylist)

## Use the clear() method to empty the list:
## mylist.clear()
## print(mylist)

## Use the reverse() method to reverse the order of the list:
mylist.reverse()
print(mylist)

## Use the sort() or sorted() method to sort the list:
mylist2 = [5, 2, 9, 1, 5, 6]
mylist2.sort() 
print(mylist2)

## You can repeat items in a list using the * operator:
mylist3 = [["banana"] * 3, ["apple"] * 2, "cherry",  True, 23] 
print(mylist3)

## You can join two lists using the + operator:
mylist4 = mylist2 + mylist3
print(mylist4)

## Use slicing to return a range of items in a list:
print(mylist2[2:5])  # Items from index 2 to 4
print(mylist2[0::2])   # :: step index
print(mylist2[::-1]) # Negative indexing to reverse the list

## Copy a list, you can use the copy() method to make a copy of a list or the slice operator [:], returns a copy of the list.
fruit=["apple", "banana", "cherry", "date"]
fruit_copy = list(fruit)  # Using list() constructor to copy a list
fruit_copy.append("elderberry") ## Modifying the copied list
print(fruit_copy)

## To have the squared values of a list:
numbers = [1, 2, 3, 4, 5]
squared_numbers = [x**2 for x in numbers]
print(squared_numbers)
```

```python
## A tuple is a collection which is ordered and unchangeable. similar to lists but you cannot modify them after creation.
mytuple = tuple(["Mary", 25, "Paris",])
print(mytuple)

## To access items in a tuple, you can use indexing just like in lists:
item = mytuple[1]
print(item)  # Output: 25

 ##print(mytuple[0]="John")  # This will raise an error because tuples are immutable 

for t in mytuple:
        print(t)

if "Asamoah" in mytuple:
        print("Yes")
else:
        print("No")


mytuple2 = ("a", "p", "p", "l", "e")    
print((mytuple2))  # Output: 5
print(mytuple2.count("p"))  # Output: 2
print(mytuple2.index("l"))  # Output: 3, the index of the first occurrence of "l"

## Convert a tuple to a list using the list() constructor:
mylist_from_tuple = list(mytuple2)
print(mylist_from_tuple)

## Convert a list to a tuple using the tuple() constructor:
mytuple_from_list = tuple(mylist_from_tuple)
print(mytuple_from_list)

## Unpacking a tuple into variables:
Name, age, city,  = mytuple
print(Name)
print(age)
print(city)

## Unpacking with asterisk * to collect multiple items:
number_tuple = (1, 2, 3, 4, 5)
i1, *i2, i3 = number_tuple
print(i1)  # Output: 1
print(i2)  # Output: [2, 3, 4]
print(i3)  # Output: 5

```
